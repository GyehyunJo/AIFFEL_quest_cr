```python
import random

class Account:
    account_count = 0  # 클래스 변수 활용 계좌 객체 개수 저장

    def __init__(self, name, balance):  # 생성자 활용해서 은행이름, 예금주, 계좌번호 설정
        self.deposit_count = 0  # 입금 회수 저장변수 생성
        self.deposit_log = []  # 입금 기록
        self.deposit_balance_log = []  # 입금 후 잔액 기록
        self.bonus_log = []  # 이자 지급 기록
        self.withdraw_log = []  # 출금 기록
        self.name = name
        self.balance = balance
        self.bank = "SC은행"

        num1 = random.randint(0, 999)  # 계좌 번호 생성
        num2 = random.randint(0, 99)
        num3 = random.randint(0, 999999)

        num1 = str(num1).zfill(3)
        num2 = str(num2).zfill(2)
        num3 = str(num3).zfill(6)

        self.account_number = num1 + '-' + num2 + '-' + num3

        Account.account_count += 1  # 계좌 객체 개수 추가

    @classmethod
    def get_account_num(cls):  # 생성된 계좌의 개수 출력
        print(cls.account_count)

    def deposit(self, amount):  # 입금 메소드
        if amount >= 1:
            self.deposit_count += 1  # 입금 회수 체크
            self.deposit_log.append(amount)  # 입금 내용 저장

            self.balance += amount
            # 입금 후 잔액 기록
            self.deposit_balance_log.append(self.balance)

            if self.deposit_count % 5 == 0:  # 이자 지급
                bonus =int(self.balance * 0.01)
                self.balance += bonus
                self.bonus_log.append((self.deposit_count, bonus, self.balance))
                print(f"{self.deposit_count}회: 이자지급   금액:{bonus}          잔액: {self.balance}")
        else:
            print("입금은 최소 1원 이상만 가능합니다.")

    def withdraw(self, amount):  # 출금 메소드
        if self.balance >= amount:
            self.withdraw_log.append(amount)
            self.balance -= amount
        else:
            print("계좌 잔고 이상으로 출금할 수 없습니다.")

    def display_info(self):  # 정보출력 메소드
        print("은행이름: ", self.bank)
        print("예금주: ", self.name)
        print("계좌번호: ", self.account_number)
        print("잔고", self.balance)

    def withdraw_history(self):  # 출금내역 출력
        current_balance = self.balance + sum(self.withdraw_log)  # 출금하기 전 잔액
        for idx, amount in enumerate(self.withdraw_log, 1):
            current_balance -= amount  # 출금 내역 차감
            print(f"{idx}회: 출금         금액: {amount}          잔액: {current_balance}")

    def deposit_history(self):    # 입금내역 출력
        events = []       #입금, 이자 발생 내역 저장

        for idx, amount in enumerate(self.deposit_log, 1):           #입금 내역 불러오기
            balance_after_deposit = self.deposit_balance_log[idx - 1]
            events.append((idx, '입금', amount, balance_after_deposit))

        for idx, (deposit_count, bonus, balance_after_bonus) in enumerate(self.bonus_log, 1):       #이자지급 내역 불러오기

            events.append((deposit_count, '이자지급', bonus, balance_after_bonus))

        events.sort(key=lambda x: x[0])         #거래 순서에 따라 정리

        for idx, event_type, amount, balance in events:    #출력
            if event_type == '입금':
                print(f"{idx}회: 입금       금액: {amount}          잔액: {balance}")
            elif event_type == '이자지급':
                print(f"{idx}회: 이자지급   금액:{amount}          잔액: {balance}")

# 계좌 생성 및 테스트
data = []  # 3개 이상 인스턴스 생성, 리스트 저장
first_account = Account("차정은", 1000)
second_account = Account("박광석", 3000000)
third_account = Account("조웅제", 50000000)

data.append(first_account)
data.append(second_account)
data.append(third_account)

for c in data:  # 잔고가 100만원 이상인 고객만 출력
    if c.balance >= 1000000:
        c.display_info()
        print("")

# 테스트: 입금과 출금 내역
for i in range(10):
    first_account.deposit(1000)
first_account.deposit(-1000)
first_account.withdraw(3000000)
first_account.withdraw(3000)

print("")
first_account.deposit_history()
print('-' * 50)
first_account.withdraw_history()


```

```python
import random

class Character:            #캐릭터 클래스 만들기
    def __init__(self, name, level, health, attack_power, defense):
        self.name = name
        self.level = level
        self.health = health
        self.attack_power = attack_power
        self.defense = defense

    def is_alive(self):     #생존 확인
        return self.health > 0

    def take_damage(self, damage):      #피격 데미지 계산
        reduced_damage = max(damage - self.defense, 0)    #방어력보다 공격력이 약하면 공격 무효
        self.health -= reduced_damage

    def attack_target(self, target):        #공격 함수
        damage = random.randint(1, self.attack_power)
        print(f"{self.name}이 {target.name}에게 {damage} 만큼 공격했다...!")
        target.take_damage(damage)
        print(f"{target.name}의 체력: {target.health}")

class Player(Character):      #캐릭터를 상속받은 플레이어
    def __init__(self, name):   #캐릭터 값 초기화
        super().__init__(name, 1, 100, 25, 5)
        self.experience = 0

    def gain_experience(self, amount):    #경험치 획득
        self.experience += amount
        print(f"{self.name}이 {amount} 경험치를 얻었습니다!")

    def level_up(self):     #레벨업
        if self.experience >= 50:
            self.experience -= 50
            self.level += 1
            self.attack_power += 10
            self.defense += 5
            print(f"{self.name}이 레벨업했습니다! 현재 레벨: {self.level}")

class Monster(Character):     #캐릭터를 상속받은 몬스터 생성
    def __init__(self, name, level):
        health = random.randint(10, 30) * level
        attack_power = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, health, attack_power, defense)

def battle(player, monster):    #전투
    print("")
    print(f"{monster.name}과의 전투를 시작합니다.")
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)
        if not player.is_alive():
            print("전투 패배..")
            return
    if player.is_alive():
        print(f"{monster.name}을 이겼다!")
        player.gain_experience(monster.level * 20)
        player.level_up()


def main():     #메인 함수
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}

    player_name = input("이름을 입력하세요: ")
    player = Player(player_name)

    for monster_name, level in monster_dict.items():    #딕셔너리 키와 값 반복
        monster = Monster(monster_name, level)
        battle(player, monster)
        if not player.is_alive():
            print("게임오버")
            return
    print("모든 몬스터를 처치했습니다!")

main()
```
