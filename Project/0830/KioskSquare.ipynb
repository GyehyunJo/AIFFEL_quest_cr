```python
# 문제 3-3. 클래스 업데이트
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요!
# 문제 3-3. 클래스 업데이트
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요!

class Kiosk:
    def __init__(self):    #생성자
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        self.order_menu = []              #메뉴와 가격 저장할 리스트
        self.order_price = []

        n=0                              #음료 정하기
        while n < 1 or len(self.menu) <n:
            print("")
            n = int(input("음료 번호를 입력하세요 : "))

            if 1 <= n and n <= len(self.menu):
                self.menu_cho=self.menu[n-1]
                self.price_cho=self.price[n-1]

            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        t=0                                     #온도정하기
        while t!=1 and t != 2:
            t = int(input ("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

            if t==1:
                self.temp='HOT'

            if t==2:
                self.temp='ICE'

            elif t!=1 and t!=2:
                print("1과 2 중 하나를 입력하세요.\n")

        self.order_menu.append(f'{self.temp} {self.menu_cho}')       #주문 리스트에 추가
        self.order_price.append(self.price_cho)

        print(f'선택하신 메뉴는 {self.temp} {self.menu_cho} 가격은 {self.price_cho}입니다.')

        while n!=0:            #지불로 넘어갈때까지 추가 주문
            print()
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))
            if n > 0 and n < len(self.menu) +1:
                temp = None
                while temp not in ['HOT', 'ICE']:
                    t = int(input("HOT 음료는 1, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        temp = 'HOT'
                    elif t == 2:
                        temp = 'ICE'
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                menu_cho = self.menu[n-1]
                price_cho = self.price[n-1]

                self.order_menu.append(f'{temp} {menu_cho}')
                self.order_price.append(price_cho)

                print(f'추가 주문 음료는 {temp} {menu_cho} : {price_cho}원\n',
                f'현재 합계 : {sum(self.order_price)}원')

            elif n == 0:   #지불 선택시
                print("")
                print("주문이 완료되었습니다.")
                print(f'주문 리스트: {self.order_menu}')
                print(f'가격 리스트: {self.order_price}')
                print(f'총 합계: {sum(self.order_price)}원')
                break
        else:
            print("없는 메뉴입니다. 다시 주문해 주세요.")

    # 지불
    def pay(self):
        print(f'지불할 금액: {sum(self.order_price)}원')

        while True:     #지불 방법 선택
            payment_method =input("지불 방법을 선택하세요 (현금은 1, 카드 결제는 2): ")

            if payment_method == '1' or payment_method.lower() == "cash":
                print('직원을 호출하겠습니다.')
                break
            elif payment_method == '2' or payment_method.lower() == 'card':
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
                break
            else:
                print("1이나 2 또는 'cash'와 'card' 중 하나를 입력하세요. 다시 결제를 시도해 주세요.")

    # 주문서 출력
    def table(self):
        print('⟝' + '-' * 30 + '⟞')    #주문서 모양내기
        for i in range(5):
            print('|' + ' ' * 31 + '|')


        for i in range(len(self.order_menu)):  # 주문 상품명 : 해당 금액
            print(self.order_menu[i], ' : ', self.order_price[i])

        print(f'합계 금액 : {sum(self.order_price)}')


        for i in range(5):    # 외곽 모양내기
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')

a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```

```python
# 사각형 넓이를 구하는 클래스 완성!
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')
            print(self.rect())


        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
            print(self.par())

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
            print(self.trape())

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):                                     #직사각형 넓이 구하기
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):                                     #평행사변형 넓이 구하기
        width, vertical = map(int, input('밑변, 높이를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '평행사변형의 넓이는 : ' + str(area)
        return result


    def trape(self):                                    #사다리꼴 넓이 구하기
        top_width, bottom_width, height = map(int, input('윗변, 아랫변, 높이를 입력하세요. 예시 : 윗변,아랫변,높이\n >>>').split(','))
        area = 0.5 * (top_width + bottom_width) * height
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result

a = Square()  # 객체 생성 & 2, 3을 각각 입력해 봅시다.
```
